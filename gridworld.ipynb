```python
import numpy as np
import gymnasium as gym
```

```python
class GridWorldEnv(gym.Env): 
    def __init__(self, 
                 size: int = 10,
                 num_types_special_regions: int = 0): 
        self.size = size

        # default un-initialized locations
        self._agent_location = np.array([-1, -1], dtype=np.int32)
        self._target_location = np.array([-1, -1], dtype=np.int32)
        self._special_regions = np.array([[] for special_region in range(num_types_special_regions)], dtype=object) # each index corresponds to a different type of region, and the values in the list are each tuples of coordinates where the region is

        self.action_space = gym.spaces.Discrete(4) # up, down, left, right

        # full agent action -> environment interpretation wrapper would be overkill, so just simple lookup table from tutorial
        self._action_to_direction = {
            0: np.array([0, 1]), # right (col + 1)
            1: np.array([-1, 0]), # up (row - 1)
            2: np.array([0, -1]), # left (col - 1)
            3: np.array([1, 0]), # down (row + 1)
        }

        def _get_obs(self): 
            return {"agent": self._agent_location, "target": self._target_location}
```
